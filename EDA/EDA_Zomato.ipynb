```python
# merge is a function that helps to combine the dataframe
final_df=pd.merge(df,df_country,on='Country Code',how= 'left')
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## Pie Chart w top 3 country using zomagto

plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Zomato maximum records/transcations are from India.Second highest is USA and third is UK. 

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observations:

1. Whenever the rating is between 4.5 t 4.9 --> Excellent
2. Whenever the rating is between 4.0 to 4.5 --> Very Good
3. Whenever the rating is between 3.5 to 3.9 --> Good
4. Whenever the rating is between 3.0 to 3.4 --> Average
5. Whenever the rating is between 2.5 to 2.9 --> Average
6. Whenever the rating is between 2.0 to 2.4 --> Poor
7. Whenever the rating is between 0.0 to 1.9 --> Poor


```python
ratings.head()
```

```python
import matplotlib 
matplotlib.rcParams['figure.figsize']= (12,6)
sns.barplot(x= 'Aggregate rating', y='Rating Count', data=ratings,hue= 'Rating color',palette=['blue','red','orange','yellow','green','green'])
```

### Observation:
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
### Count plot

sns.countplot(x= 'Rating color', data= ratings, palette= ['blue','red','orange','yellow','green','green'])
```
